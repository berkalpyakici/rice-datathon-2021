```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/Cleaned_Purchase_Card_Transactions.csv")
```

```python
df.columns
```

```python
all_payments = []
for agency_num in range(89):
    df_agency = df[df.AGENCY_NUM == agency_num]
    payments = {int(vendor): 0 for vendor in np.unique(df_agency.VENDOR_NAME_NUM)}
    
    # Add payments for vendor
    for vendor in np.unique(df_agency.VENDOR_NAME_NUM):
        df_vendor = df_agency[df_agency.VENDOR_NAME_NUM == vendor]
        payments[vendor] += float(np.sum(df_vendor.TRANSACTION_AMOUNT))
    all_payments.append(payments)
all_payments
```

```python
import json

with open("data/payments.json", "w") as f:
    f.write(json.dumps(all_payments))
```
